```python
import requests
import json

class king(object):
    def __init__(self,word):
        self.url = 'https://www.iciba.com/translate'
        self.headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36 Edg/143.0.0.0'}
        self.data = {
          "from": "auto",
          "to": "auto",
          "textList": [word],
        }
    def get_data(self):
        response = requests.post(self.url,headers=self.headers,data=self.data)
        print("状态码:", response.status_code)
        print("响应头:", dict(response.headers))
        print("响应内容:", response.text)
        print("响应URL:", response.url)
        return response.content
    def run(self):
        response = self.get_data()
        print(response.decode('utf-8'))

f = king("苦难")
f.run()
```

```python
import requests
import json

def translate_google(text, src='auto', dest='zh-CN'):
    url = "https://translate.googleapis.com/translate_a/single"
    params = {
        "client": "gtx",
        "sl": src,
        "tl": dest,
        "dt": "t",
        "q": text
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        result = response.json()
        return ''.join([item[0] for item in result[0]])
    else:
        return None

print(translate_google("what a sorrow day !"))
```
